Будуємо гістограму

```python
plt.subplots(figsize=(10,8))
plt.hist(proba, density=True)
```

Будуємо криву емпіричної функцію розподілу

```python
N=len(proba)
X = np.sort(proba)
F = np.array(range(N))/float(N)


plot.plot(X, F)
plot.show()
```

Знайдемо параметри нормального розподілу

```python
a, sigma = scipy.stats.distributions.norm.fit(proba)
a, sigma
```

Отже, майже всі значення знаходяться у мроміжку між -0.138 до 0.984, з середнім значенням 0.423

Отримуємо два параметри - середнє та середньє квадратичне відхилення

Тепер побудуємо криву нормального розподілу

```python
ix = np.linspace(0,1,20)
```

```python
N_fitted_proba = scipy.stats.distributions.norm.pdf(ix, a, sigma)

```

```python
plt.hist(proba, density=True)
plt.plot(ix,N_fitted_proba,'g')
```

Припустимо, що вибірка має якийсь розподіл Стьюдента. Знайдемо його і побудувати криву.

```python
x, loc, scale  =scipy.stats.distributions.t.fit(proba)
```

Маємо параметри

```python
x, loc, scale
```

```python
Ox = np.linspace(0,1,15)
```

```python
t_fitted_proba = scipy.stats.distributions.t.pdf(Ox, x, loc, scale)
```

```python
plt.subplots(figsize=(10,6))
plt.hist(proba, density=True)

plt.plot(Ox,t_fitted_proba,'r')
```
